```python
import pandas as pd
import datetime
import numpy as np
```

```python
def separateCityState(d):
    d['City'] = ""
    for index, row in d.iterrows():
        if ', ' in str(row['Province/State']):
            split = str(row['Province/State']).split(", ")
            city = split[0]
            state = split[1]
            d.loc[index, 'City'] = city
            d.loc[index, 'Province/State'] = state
    return d

def replaceCountry(d):
    for index, row in d.iterrows():
        if str(row['Country/Region']) == "Hong Kong" or str(row['Country/Region']) == "Mainland China":
            d.loc[index,'Country/Region'] = 'China'
    return d

def trimDate(d):
    for index, row in d.iterrows():
        t = str(row['Last Update']).split(" ")[0]
        d.loc[index, 'Last Update'] = t.split("T")[0]
        
    return d
```

```python
# make CSV and put first day of data in
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-22-2020.csv")
df['Latitude'] = ""
df['Longitude'] = ""
df = separateCityState(df)
df = df.replace(np.nan, '', regex=True)
df = replaceCountry(df)
df = trimDate(df)
df.to_csv("AggregatedDailyData.csv")
df.head()
```

```python
# append the rest of the days of data
day_delta = datetime.timedelta(days=1)
end_date = datetime.date.today()
start_date = datetime.date(2020,1,23)
for i in range((end_date - start_date).days):
    date = (start_date + i*day_delta).strftime("%m-%d-%Y")
    print(date)
    df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/" + date + ".csv")
    if "Latitude" not in df.columns:
        df['Latitude'] = ""
        df['Longitude'] = ""
    df = separateCityState(df)
    df = replaceCountry(df)
    df = trimDate(df)
    df.to_csv("AggregatedDailyData.csv", mode='a', header=False)
```
